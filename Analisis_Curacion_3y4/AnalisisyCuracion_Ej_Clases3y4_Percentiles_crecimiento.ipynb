####   Graficar las curvas correspondientes a cada percentil para niños y niñas

El dataset se obtiene de https://www.cdc.gov/growthcharts/percentile_data_files.htm
Para este práctico se tratan los dataset 1 y 2


```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Lectura de datasets
# https://www.cdc.gov/growthcharts/percentile_data_files.htm
# datasets 1 y 2
weight_for_age = pd.read_csv("https://raw.githubusercontent.com/fpalaciosdrobins/diplodatos/master/Analisis_Curacion_3y4/wtageinf.csv")

weight_for_age = weight_for_age[['Sex','Agemos','P3','P5','P10','P25','P50','P75','P90','P95']]

weight_for_age.describe(include='all')
```

```python
weight_for_age_sex_1 = weight_for_age.loc[weight_for_age['Sex'] == 1]
```

### Curvas correspondientes a Pesos para la Edad

```python
plt.figure(figsize=(10,5))
plt.title('Peso para la Edad en Niños')
plt.xlabel('Edad en Meses')
plt.ylabel('Peso en Kg')
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Set1')

data_reducido = weight_for_age_sex_1
num=0
for column in weight_for_age_sex_1.drop('Sex', axis=1):
    num+=1
    if num>1:
        plt.plot(weight_for_age_sex_1["Agemos"], weight_for_age_sex_1[column], marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)
        
# Add legend
plt.legend(loc=2, ncol=1)
```

```python
len_for_age = pd.read_csv("https://raw.githubusercontent.com/fpalaciosdrobins/diplodatos/master/Analisis_Curacion_3y4/lenageinf.csv")
len_for_age.describe()

len_for_age = len_for_age[['Sex','Agemos','P3','P5','P10','P25','P50','P75','P90','P95']]
```

### Curvas correspondientes al Largo para la Edad

```python
plt.figure(figsize=(10,5))
plt.title('Altura para la Edad en Niños')
plt.xlabel('Edad en Meses')
plt.ylabel('Altura en cm')
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Set1')

# Es necesario convertir a numérico las columnas

num=0
for column in len_for_age_sex_1.drop('Sex', axis=1):
    num+=1
    if num>1:
        plt.plot(len_for_age_sex_1["Agemos"], pd.to_numeric(len_for_age_sex_1[column], errors='ignore'), marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)
        
# Add legend
plt.legend(loc=2, ncol=1)
```
